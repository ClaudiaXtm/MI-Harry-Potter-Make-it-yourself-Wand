```python
#Split video into frames
import cv2
import glob
import numpy as np

# Opens the Video file
cap= cv2.VideoCapture('./modifiedAM/modified-revelio25.avi')
i=0
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    cv2.imwrite('./framesAM/frame'+str(i)+'.jpg',frame)
    i+=1
 
cap.release()
cv2.destroyAllWindows()

#Add frames to form shape
images = [cv2.imread(file) for file in glob.glob('./framesAM/*jpg')]

i=0
img = 0
for i in range(len(images)):
    if(i<len(images)):
        img = cv2.add(img, cv2.add(images[i], images[i+1]))
        cv2.imwrite('./merged_frames/revelio25.jpg',img)

# Show the image 
cv2.imshow('image', img) 
  
# Wait for a key 
cv2.waitKey(0) 
  
# Destroy all the window open 
cv2.destroyAllWindows()

# Adjust image colors and sizes
img = cv2.imread('./merged_frames/revelio25.jpg')

#apply dilation to increase the shape
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 2)
cv2.imshow('Dilation', dilation)

#save dilated photo
cv2.imwrite('./merged_frames/dilated/revelio25.jpg',img)
print( 'Original Dimensions: ', img. shape)

#define size
width = 28
height = 28
dim = (width, height)

# resize image
resized = cv2.resize(dilation, dim, interpolation = cv2.INTER_AREA)
print('Resized Dimensions : ',resized.shape)
cv2.imshow('Resized', resized)

#save processed photo
cv2.imwrite('./merged_frames/cropped/Revelio025.jpg',resized)
# Wait for a key 
cv2.waitKey(0) 
  
#Destroy all the window open 
cv2.destroyAllWindows()

```
